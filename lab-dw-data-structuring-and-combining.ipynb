```python
import pandas as pd
df_1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
df_2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')

df_1.head()
```

```python
df_1.columns = [col.lower() for col in df_1.columns]
df_1.columns
```

```python
df_2.columns = [col.lower() for col in df_1.columns]
df_2.columns
```

```python
df_1.columns = df_1.columns.str.replace(' ','_')
df_1.columns
```

```python
df_2.columns = df_2.columns.str.replace(' ','_')
df_2.columns
```

```python
df_1.rename(columns={'st': 'state'}, inplace=True)
```

```python
df_2.rename(columns={'st': 'state'}, inplace=True)
```

```python
df_2
```

```python
df_2['gender'],df_2['customer_lifetime_value'] = df_2['customer_lifetime_value'].copy(),df_2['gender'].copy()
```

```python
df_2
```

```python
df_2['total_claim_amount'],df_2['policy_type'] = df_2['policy_type'].copy(),df_2['total_claim_amount'].copy()
```

```python
df_2
```

```python
df = pd.concat([df_1,df_2],ignore_index=True)
df
```

```python

gender_dict ={
    'F':'F',
    'female':'F',
    'Femal':'F',
    'M':'M',
    'Male':'M',
    
}
```

```python
df['gender'] = df['gender'].map(gender_dict)
```

```python
state_dict = {
    'Washington':'Washington',
    'Arizona':'Arizona',
    'Nevada':'Nevada',
    'California':'California',
    'Oregon':'Oregon',
    'Cali':'California',
     'AZ':'Arizona',
     'WA':'Washington',

}
df['state'] = df['state'].map(state_dict)

```

```python
df = df.rename(columns={'bachelors':'bachelor'})
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%','')
df['vehicle_class'] = df['vehicle_class'].replace({
    'Luxury SUV':'Luxury',
    'Sports Car':'Luxury',
    'Luxury Car':'Luxury'
})
```

```python
df[df['customer_lifetime_value'].isna()]
```

```python
#converted customer lifetime value column into float type
df['customer_lifetime_value'] = df['customer_lifetime_value'].astype(float)
df['customer_lifetime_value']
```

```python
#fixing the changes
df['number_of_open_complaints'] = df['number_of_open_complaints'].apply(lambda x: x.split('/')[1] if type(x) == str else x )
```

```python
df['number_of_open_complaints'] = df['number_of_open_complaints'].astype(float)
df['number_of_open_complaints'].unique()
```

```python
df.isna().sum()
```

```python
df_cleaned = df.dropna(subset='customer')
```

```python
df_cleaned
```

```python
#used bfill method to fill the nan values as the first one itself is nan value
df_cleaned['gender'] = df_cleaned['gender'].fillna(method='bfill')
```

```python
#used mean to fill the nan values here to keep it in average
df_cleaned['customer_lifetime_value'] = df_cleaned['customer_lifetime_value'].fillna('mean')
```

```python
df_cleaned.isna().sum()
```

```python
#checking if there is duplicates in cleaned data frame
df_cleaned.duplicated().value_counts()
```

```python
# Drop duplicate rows
df_cleaned.drop_duplicates(inplace=True)
```

```python
pivot_table = df2.pivot_table(
    index='gender',
    columns='education',
    values='customer_lifetime_value',
    aggfunc='mean'
).round(2)
print(pivot_table)
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
df2.columns
```

```python
df2['effective_to_date'] = pd.to_datetime(df2['effective_to_date'])
df2['month'] = df2['effective_to_date'].dt.month_name()
complaints_summary = df2.groupby(['policy_type','month']).size().reset_index(name='Number_Of_Complaints')
complaints_summary = complaints_summary.sort_values(by=['policy_type','month'])
print(complaints_summary)
```
